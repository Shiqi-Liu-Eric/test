## 4. Alpha Factory

The Alpha Factory is the research engine responsible for discovering, validating, and cataloging tradeable signals.

```
Research Hypothesis
       │
       ▼
  Signal Coding (alpha_xxxx_NN.py)
       │
       ▼
  Backtesting (backtest_xxxx_NN.xml + superRun.py)
       │
       ▼
  Risk Neutralization (Barra / Market / Industry)
       │
       ▼
  Correlation Check (vs. existing alpha pool)    --->  Reject if abs(corr) too large
       │
       ▼
  Paper Trading (shadow execution for N days)
       │
       ▼
  Production Deployment (strategy_main/strategy_alphaCombo)
       │
       ▼
  Ongoing Monitoring (decay tracking, IC analysis)
       │
       ▼
  Retirement (when IC decays below threshold)
```

Each alpha is evaluated on the following metrics before promotion (not limited to these):
- **Information Coefficient (IC)**: Rank correlation between predicted and realized returns
- **IC Information Ratio (ICIR)**: IC mean / IC standard deviation
- **Turnover**: Daily portfolio turnover rate (lower is better for cost)
- **Sharpe Ratio**: Risk-adjusted return (annualized, net of estimated transaction costs)
- **Max Drawdown**: Peak-to-trough decline
- **Sharpe Ratio**: $SR = \frac{R_p - R_f}{\sigma_p}$ | $R_p$: portfolio return, $R_f$: risk-free rate, $\sigma_p$: portfolio volatility (annualized)
- **Calmar Ratio**: $CR = \frac{R_{\text{annual}}}{\text{MaxDD}}$ | $R_{\text{annual}}$: annualized return, MaxDD: maximum drawdown (absolute value)
- **Information Ratio**: $IR = \frac{R_p - R_b}{\sigma_{p-b}}$ | $R_b$: benchmark return, $\sigma_{p-b}$: tracking error (std of excess returns)
- **Decay Half-Life**: Days until signal IC drops to 50% of original
- **Correlation to Existing Pool**: Must be below threshold to add diversification value

## 5. Backtesting Engine

### 5.1 Overview
The backtesting engine is built around an XML-templated configuration system that separates **strategy logic** (alpha signals) from **execution parameters** (universe, rebalance frequency, costs, constraints). This enables rapid iteration without modifying code.

### 5.2 XML Configuration Example (may differs from PROD mode .xml code)

```xml
<backtest>
    <meta>
        <name>alpha_pv_momentum_01</name>
        <author>Shiqi Liu</author>
        <created>2025-11-25</created>
    </meta>
    <universe>SP500</universe>
    <frequency>daily</frequency>
    <lookback_years>5</lookback_years>
    <alpha_source>alpha_pv_01.py</alpha_source>
    <neutralization>
        <barra>true</barra>
        <market>true</market>
        <industry>true</industry>
    </neutralization>
    <constraints>
        <long_only>false</long_only>
        <max_weight>0.05</max_weight>
        <max_turnover>0.3</max_turnover>
        <max_sector_exposure>0.15</max_sector_exposure>
    </constraints>
    <costs>
        <commission_bps>1.0</commission_bps>
        <slippage_bps>2.0</slippage_bps>
        <market_impact_model>sqrt</market_impact_model>
    </costs>
    <output>
        <equity_curve>true</equity_curve>
        <group_returns>true</group_returns>
        <factor_exposure>true</factor_exposure>
    </output>
</backtest>
```

### 5.3 Backtesting Utilities
Here are some primary backtesting util scripts. They are used only on alpha backtesting.

| Utility | Function |
|---|---|
| `neut_barra` | Neutralize against Barra risk factors (size, value, momentum, volatility, etc.) |
| `neut_market` | Remove market beta exposure |
| `neut_industry` | Neutralize industry/sector tilts |
| `func_decay` | Apply exponential or linear decay weighting to signals |
| `func_delay` | Simulate T+1, T+2 execution delays |
| `func_longOnly` / `func_shortOnly` | Constrain portfolio to long-only or short-only |
| `vis_main` | Plot equity curves, drawdown charts, rolling Sharpe |
| `vis_byGroup` | Quantile-sorted portfolio group returns (Q1–Q5) |
| `print_corr` | Cross-alpha correlation heatmap |
| `print_stat` | Summary statistics table (Sharpe, Sortino, Calmar, max DD, turnover) |
| `print_log` | Detailed backtest execution log |

### 5.4 Running a Backtest

```bash
# Single alpha backtest
python backtest_main/run.py --config backtest_main/backtest_pv_01.xml

# Multi-alpha combination
python backtest_main/run.py --config backtest_main/backtest_global.xml

# Batch run all configs
python backtest_main/run.py --batch backtest_main/*.xml --parallel 4
```

### 6.2 Portfolio Construction

The portfolio optimizer (`optimizer.py`) solves the following convex program:

$$
\max_w \quad \alpha^T w - \frac{\lambda}{2} w^T \Sigma w - \gamma \|w - w_{\text{prev}}\|_1
$$

subject to:

$$
\begin{aligned}
& \sum_i w_i = 1 \quad \text{(fully invested)} \\
& |w_i| \leq w_{\max} \quad \forall i \quad \text{(position limits)} \\
& |B^T w| \leq e_{\max} \quad \text{(factor exposure limits)} \\
& \|w - w_{\text{prev}}\|_1 \leq \tau_{\max} \quad \text{(turnover constraint)}
\end{aligned}
$$

Where:
- $w$ = portfolio weights vector
- $\alpha$ = expected return vector (from alpha factory)
- $\Sigma$ = covariance matrix (Barra or sample-based)
- $\lambda$ = risk aversion parameter
- $\gamma$ = turnover penalty
- $B$ = factor loading matrix
- $e_{\max}$ = maximum factor exposure
- $\tau_{\max}$ = maximum single-period turnover

## 11. Fee Structure & Investor Terms (Pending to add ..)

### 11.1 Fee Schedule

We aim to build a multi-asset, multi-strategy portfolio, so overall performance is based on each individual strategy. We expect the overall portfolio to have 1% annualized return on 1% annualized volatility (~= 1 sharpe ratio). Current target volatility is ~10% (leverage applied) with <10% max-drawdown.

| Fee Type | Rate | Description |
|---|---|---|
| **Management Fee** | 0.2% annually | Charged on committed capital; accrued daily, collected quarterly |
| **Entry Fee** | 0.1% (one-time) | Charged on initial subscription amount at time of investment |
| **Performance Fee** | 20% of excess return | Charged on returns above the hurdle rate (see below) |
| **Underlying Costs** | Pass-through | ETF expense ratios, broker commissions, and exchange fees |

### 11.2 Performance Fee Waterfall

**For holding periods ≥ 1 year:**
- Hurdle rate: **8% annualized**
- Performance fee: **20%** of returns exceeding the hurdle rate
- High-water mark: Yes — no performance fee until prior peak NAV is recovered

**For holding periods < 1 year:**
- No hurdle rate applies
- Performance fee: **20%** of total positive returns
- If returns are negative, **no performance fee** is charged

### 11.3 Subscription & Redemption

| Term | Detail |
|---|---|
| **Minimum Investment** | USD 1,000 per individual investor |
| **Maximum Investment** | USD 10,000 per individual investor |
| **Subscription** | Accepted on any business day; NAV calculated T+0 |
