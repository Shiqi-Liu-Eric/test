## Market Basics

- **Market Participants**: brokers, proprietary traders, producers, end users, fund managers, market makers (MMs), etc.

- **订单类型**：
    - Limit：挂在市场等
    - IOC：立刻取消
    - Market：不管价格全部成交
    - FOK：必须全部立刻成交，否则取消
    - GFD：当天有效，收盘取消
    - GTC：一直有效直到成交或手动取消
    - AON：All-or-None

## Options Basics

- **Call**：买权，right to buy at strike K
- **Put**：卖权，right to sell at strike K
- American可以在到期前任意时间行权，European只能到期日行权
- 买方付premium，卖方收premium承担义务

Moneyness:
| | Call | Put |
|---|---|---|
| ITM | S > K | S < K |
| ATM | S ≈ K | S ≈ K |
| OTM | S < K | S > K |

Intrinsic value = max(0, payoff if exercised now)  
Time value = Option price - Intrinsic value，永远 ≥ 0

## Put-Call Parity

$$C - P = S - K \cdot e^{-rT}$$

直觉：long call + short put = synthetic long forward  
violation就是arbitrage机会（实际中要考虑transaction cost、dividends、early exercise）

对于American options，parity变成inequality：
$$S - K \leq C - P \leq S - K \cdot e^{-rT}$$

## Greeks

**Delta (Δ)**
- Call: 0 到 1，Put: -1 到 0
- ATM delta ≈ ±0.5
- 也可以近似理解为到期ITM的概率
- delta hedging：hold Δ shares per option来neutralize方向性风险

**Gamma (Γ)**
- delta对underlying price的导数，$\Gamma = \frac{\partial \Delta}{\partial S}$
- ATM时gamma最大，离到期越近ATM gamma越尖（pin risk）
- long options = long gamma = 喜欢大波动
- gamma scalping：delta hedge后靠realized vol > implied vol赚钱

**Theta (Θ)**
- 时间衰减，每天亏多少（long options θ < 0）
- ATM theta最大，接近到期加速
- Gamma和Theta是对立的：long gamma必然付theta，short gamma收theta
- 核心trade-off：$\Gamma \cdot \frac{\sigma^2 S^2}{2} + \Theta \approx 0$（BSM下）

**Vega (ν)**
- 对implied vol的敏感度
- ATM longest dated vega最大
- vol trading本质就是trading vega

**Rho (ρ)**
- 对利率敏感，一般影响小除非long-dated options

## Volatility

**Historical (Realized) Vol**
- 过去return的std dev，annualized: $\sigma_{ann} = \sigma_{daily} \times \sqrt{252}$
- 后视的，只能告诉你过去发生了什么

**Implied Vol (IV)**
- 从option market price反推出来的vol（BSM反解）
- 前瞻的，反映市场对未来波动的预期/定价
- 本质上options trading就是在trading implied vol vs你对realized vol的预期

**Vol Smile / Skew**
- IV不是flat的，画出来对strike是U型（smile）或者左偏（skew）
- Equity通常有negative skew（OTM puts贵）→ crash protection demand
- Commodities更接近smile
- skew的term structure也有信息：近月skew通常更陡

**Vol Surface**
- IV = f(K, T)，strike × expiry的二维surface
- sticky strike vs sticky delta：vol surface怎么随S移动的两种regime

## Option Pricing — BSM

$$C = S \cdot N(d_1) - K e^{-rT} N(d_2)$$
$$d_1 = \frac{\ln(S/K) + (r + \sigma^2/2)T}{\sigma\sqrt{T}}, \quad d_2 = d_1 - \sigma\sqrt{T}$$

假设：log-normal price distribution, constant vol, no dividends, continuous trading, no transaction cost

局限：fat tails / jumps / vol不constant → 这就是为什么有vol smile  
实际中BSM更多是一个quoting convention（用IV）而不是真正的pricing model

## Common Strategies

**Vertical Spreads**
- Bull call spread: long C(K1) + short C(K2)，K1 < K2，limited risk/reward
- Bear put spread: long P(K2) + short P(K1)
- 本质是对方向的bet，spread controls成本和风险

**Straddle & Strangle**
- Straddle: long ATM call + ATM put → bet on大波动，方向不care
- Strangle: long OTM call + OTM put → 便宜但需要更大move才能profit
- 做straddle/strangle就是在bet realized vol > implied vol

**Butterflies & Condors**
- Butterfly: long 1 C(K1) + short 2 C(K2) + long 1 C(K3) → bet on到期pin在K2
- Iron condor: short strangle + long further OTM wings → range-bound bet，收theta

**Calendar Spread**
- long longer-dated + short shorter-dated（same strike）
- bet on term structure / 近月vol相对远月太高

**Risk Reversal**
- long OTM call + short OTM put（或反过来）
- 表达skew view，也是synthetic forward的变体

## Market Making

- MM的角色：提供liquidity，bid-ask spread赚钱
- 不take方向性bet，靠delta hedge保持neutral
- 利润来源：bid-ask spread - hedging cost
- inventory management很关键，position limits

**MM怎么定价：**
1. 算出theoretical value（基于自己的vol estimate）
2. 在theo两边挂bid/ask，spread要cover hedging cost + profit margin
3. 根据inventory / risk调整skew（偏一边挂更aggressive）

**Edge的来源：**
- 更好的vol prediction（你的vol model vs 市场的IV）
- 更快的speed（低延迟抢单）
- 更好的risk management

## P&L Attribution

- Delta P&L: $\Delta \cdot \delta S$
- Gamma P&L: $\frac{1}{2}\Gamma \cdot (\delta S)^2$
- Theta P&L: $\Theta \cdot \delta t$
- Vega P&L: $\nu \cdot \delta\sigma$

Delta-hedged portfolio的P&L主要来自gamma vs theta的battle：
$$P\&L \approx \frac{1}{2}\Gamma S^2(\sigma_{realized}^2 - \sigma_{implied}^2)\delta t$$

→ 如果realized > implied，long gamma赚钱；反之short gamma赚钱

## Early Exercise & Dividends

- American call on non-dividend stock：永远不early exercise（time value > 0）
- 有dividend时，可能在ex-date前exercise call来capture dividend
  - 条件：dividend > remaining time value of the option
- American put：deep ITM时可能early exercise（拿cash赚interest > remaining time value）

## Misc Notes

- **Pin risk**: gamma在到期日ATM附近爆炸，delta在0和1之间跳，hedge很难
- **Assignment risk**: short American options随时可能被exercise
- **Liquidity**: bid-ask spread是真实的cost，OTM deep options流动性差
- **Margin**: selling naked options需要margin，portfolio margin比Reg-T便宜
- **Vol of vol**: vol本身也有vol，这影响wings的定价（kurtosis）
